```python
from keras.models import load_model
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
my_old_model=load_model('imporved')
```

```python
train_data=pd.read_csv('C:/Users/Som/AI Workshop/digit Recognizer data/train.csv')
test_data=pd.read_csv('C:/Users/Som/AI Workshop/digit Recognizer data/test.csv')
train_df_X=train_data.drop('label',axis=1)
X_train=train_df_X.values
X_test=test_data.values
y_train=train_data['label'].values
X_train.shape
X_train=X_train/255
X_test=X_test/255

test_data.shape
```

```python
from keras.layers import Dense
from keras.models import Sequential
from keras.optimizers import SGD
from keras.optimizers import Adam
```

```python
def create_model():
    m=Sequential()
    m.add(Dense(16,input_dim=784,activation='selu'))
    m.add(Dense(16,activation='selu'))
    m.add(Dense(10,activation='softmax'))
    return m
```

```python
model=create_model()
model.summary()
model.compile(loss='mse',optimizer=Adam(lr=0.001),metrics=['acc'])
from keras.utils import np_utils
y_1h=np_utils.to_categorical(y_train)
```

```python
from keras.models import load_model
my_old_model=load_model('myModel')
hisory1=my_old_model.fit(X_train,y_1h,epochs=150,batch_size=200)
```

```python
prediction=my_old_model.predict(X_test)
```

```python
def getMax(data):
    arr=[]
    i=1
    for row in data:
        arr.append([i,np.argmax(row)])
        i=i+1
    return arr
```

```python
output_data=getMax(prediction)
```
