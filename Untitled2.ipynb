```python
from keras.models import load_model
my_old_model=load_model('imporved')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
test_data=pd.read_csv('C:/Users/Som/AI Workshop/digit Recognizer data/test.csv')
```

```python
X_test=test_data.values
```

```python
my_old_model.predict(X_test)
```

```python
def getMax(data):
    arr=[]
    i=1
    for row in data:
        arr.append([i,np.argmax(row)])
        i=i+1
    return arr
#         print(np.argmax(row))
```

```python
output_data=getMax(my_old_model.predict(X_test))
```
