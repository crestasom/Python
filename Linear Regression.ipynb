```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pandas as pd
%matplotlib inline
```

```python
X=np.linspace(-1,1,101)
print(X)
noise=np.random.randn(*X.shape)*.33
plt.plot(noise)
# plt.plot(X,'g')
```

```python
Y=2*X+noise  #y=mx+c
Y
```

```python
plt.plot(X)
```

```python
plt.plot(Y)
```

```python
model=LinearRegression()
X=X.reshape(-1,1)
Y=Y.reshape(-1,1)
model.fit(X,Y) #trainig using dataset
m,c=model.coef_,model.intercept_
print(m,c)
```

```python
plt.plot(X,Y)
plt.plot(X,m*X+c,'r')
```

```python
model.score(X,Y)
```
