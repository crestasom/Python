```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

<b>Creating Training Data set: Input Data

```python
#Creating equal space 101 points
X=np.linspace(-1,1,101)
#Creating noise to randomize the data
noise=np.random.randn(*X.shape)*.33
```

```python
print(len(noise))
plt.plot(noise)
```

<b>Creating Dataset: Output. Output is linear to input

```python
Y=2*X+noise  #y=mx+c
```

```python
plt.plot(Y,"*")
```

```python
X=X.reshape(-1,1)
Y=Y.reshape(-1,1)
```

```python
model=LinearRegression()
model.fit(X,Y) #trainig using dataset
m=model.coef_
c=model.intercept_
print("Slope=",m,"Intercept=",c)
```

```python
plt.plot(X,Y,"*")
plt.plot(X,m*X+c,'r')
```

```python
model.score(X,Y)
```
